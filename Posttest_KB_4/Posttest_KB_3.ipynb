# Import Library yang dibutuhkan
- pandas digunakan untuk mengolah dataset ke dalam dataframe
- sklearn berguna untuk data preprocessing, diantaranya melakukan pemisahan data train dan test, pengisian nilai null, normalisasi, dan standardisasi

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

# Memasukkan dataset ke dataframe
Dataset yang akan dipakai adalah weatherAUS.csv yang berisi rincian cuaca  untuk menentukan apakah besok akan hujan atau tidak

```python
df = pd.read_csv('/content/weatherAUS.csv')
```

# Melakukan pemisahan dataset ke dalam training set dan test set
label yang akan digunakan adalah atribut RainTomorrow. Atribut sisanya akan menjadi fitur yang dipakai untuk memprediksi nilai atribut RainTomorrow

# Mencari data yang kosong

```python
df.isna().sum()
```

# Mengisi tiap data yang kosong
Pada data bertipe objek, dilakukan pengisian menurut data terbanyak, sementara pada data float dilakukan pengisian sesuai nilai mean.

```python
mean = SimpleImputer(strategy='mean')
median = SimpleImputer(strategy='median')

df['MinTemp'] = mean.fit_transform(df[['MinTemp']])
df['MaxTemp'] = mean.fit_transform(df[['MaxTemp']])
df['Rainfall'] = mean.fit_transform(df[['Rainfall']])
df['Evaporation'] = mean.fit_transform(df[['Evaporation']])
df['Sunshine'] = mean.fit_transform(df[['Sunshine']])
df['WindGustDir'].fillna(df['WindGustDir'].mode()[0], inplace=True)
df['WindGustSpeed'] = mean.fit_transform(df[['WindGustSpeed']])
df['WindDir9am'].fillna(df['WindDir9am'].mode()[0], inplace=True)
df['WindDir3pm'].fillna(df['WindDir3pm'].mode()[0], inplace=True)
df['WindSpeed9am'] = mean.fit_transform(df[['WindSpeed9am']])
df['WindSpeed3pm'] = mean.fit_transform(df[['WindSpeed3pm']])
df['Humidity9am'] = mean.fit_transform(df[['Humidity9am']])
df['Humidity3pm'] = mean.fit_transform(df[['Humidity3pm']])
df['Pressure9am'] = mean.fit_transform(df[['Pressure9am']])
df['Pressure3pm'] = mean.fit_transform(df[['Pressure3pm']])
df['Cloud9am'] = mean.fit_transform(df[['Cloud9am']])
df['Cloud3pm'] = mean.fit_transform(df[['Cloud3pm']])
df['Temp9am'] = mean.fit_transform(df[['Temp9am']])
df['Temp3pm'] = mean.fit_transform(df[['Temp3pm']])
df['RainToday'].fillna(df['RainToday'].mode()[0], inplace=True)
df['RainTomorrow'].fillna(df['RainTomorrow'].mode()[0], inplace=True)
```

# Mengecek kembali data yang kosong
Data yang bernilai null kini sudah tidak ada

```python
df.isna().sum()
```

# Membuat data duplikat pada dataset
Karena dataset berikut tidak memiliki nilai duplikat, maka disini akan dibuat terlebih dahulu nilai duplikat, dengan menyalin dataframe baris ke 2 ke dataframe kembali

```python
df.loc[len(df.index)] = df.loc[1]
```

# Mengecek jumlah data duplikat
Nilai duplikat merupakan data yang baru ditambahkan tadi

```python
df.duplicated().sum()
```

# Menghapus data duplikat

```python
df.drop_duplicates(inplace=True)
```

# Mengecek kembali jumlah data duplikat
Kini data duplikat sudah hilang

```python
df.duplicated().sum()
```

# Melakukan normalisasi pada data
Atribut yang akan dinormalisasi dan di standardisasi hanya atribut bertipe angka seperti int atau float

```python
angka = df.select_dtypes(include=['float64', 'int64']).columns

df_normal = df.copy()
df_normal[angka] = MinMaxScaler().fit_transform(df_normal[angka])

df_normal.head()
```

# Melakukan Standardisasi pada data

```python
df_standar = df.copy()
df_standar[angka] = StandardScaler().fit_transform(df_standar[angka])

df_standar.head()
```

# Mengubah tipe data salah satu kolom
Karena semua data angka bernilai float, maka disini akan dirubah satu atribut ke nilai int

```python
df['WindGustSpeed'] = df['WindGustSpeed'].astype(int)
```

# One Hot Encoding pada data
Atribut data yang akan di encoding adalah semua atribut dengan sifat kategorikal.

```python
encode = pd.get_dummies(df[['Location']])
df = pd.concat([df, encode], axis=1)
df = df.drop(columns=['Location'])

encode = pd.get_dummies(df[['WindGustDir']])
df = pd.concat([df, encode], axis=1)
df = df.drop(columns=['WindGustDir'])

encode = pd.get_dummies(df[['WindDir9am']])
df = pd.concat([df, encode], axis=1)
df = df.drop(columns=['WindDir9am'])

encode = pd.get_dummies(df[['WindDir3pm']])
df = pd.concat([df, encode], axis=1)
df = df.drop(columns=['WindDir3pm'])

encode = pd.get_dummies(df[['RainToday']])
df = pd.concat([df, encode], axis=1)
df = df.drop(columns=['RainToday'])

encode = pd.get_dummies(df[['RainTomorrow']])
df = pd.concat([df, encode], axis=1)
df = df.drop(columns=['RainTomorrow'])
```
